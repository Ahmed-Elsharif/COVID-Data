```python
import requests
```

```python
response = requests.get("http://coronavirus.m.pipedream.net/")
response
```

```python
data = response.json()
data
```

```python
data.keys()
```

```python
raw_data = data["rawData"]
raw_data
```

```python
for country in raw_data:
    name = country["Country_Region"]
    state = country["Province_State"]
    confirmed = country["Confirmed"]
    deaths = country["Deaths"]
    recovered = country["Recovered"]
    fatality = country["Case_Fatality_Ratio"]

    row = f"{name},{state},{confirmed},{deaths},{recovered},{fatality}\n"
    print(row)
```

```python
with open("covid.csv", mode="w") as f:
    f.write("Country Name,State,Confirmed,Deaths,Recovered,Case Fatality Ratio\n")

    for country in raw_data:
        name = country["Country_Region"]
        state = country["Province_State"]
        confirmed = country["Confirmed"]
        deaths = country["Deaths"]
        recovered = country["Recovered"]
        fatality = country["Case_Fatality_Ratio"]

        row = f"{name},{state},{confirmed},{deaths},{recovered},{fatality}\n"
        f.write(row)
```
